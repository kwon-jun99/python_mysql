```python
import pymysql

conn = pymysql.connect(
    host="localhost",
    user="root",
    password="class1234",
    database="test",
    charset="utf8"
)

cursor = conn.cursor() # 연결된 DB에 SQL을 실행하는 객체

print("MySQL 연결 성공")
```

```python
products = [
    {"name": "키보드", "price": 50000},
    {"name": "마우스", "price": 25000},
    {"name": "모니터", "price": 300000},
    {"name": "노트북", "price": 1200000},
    {"name": "USB", "price": 15000},
    {"name": "웹캠", "price": 80000},
    {"name": "헤드셋", "price": 90000},
    {"name": "스피커", "price": 70000},
    {"name": "외장하드", "price": 140000},
    {"name": "프린터", "price": 220000}
]
```

```python
sql = """
INSERT INTO product(name, price)
VALUES(%s, %s)
"""

for product in products:
    # execute(str, tuple) 채울 표현식%가 없으면 두번째 인수는 생략, 하나라도 있으면 튜플로 순서데로 입력
    # ex) (1,) , (1, 1), (1, 1, 1), ...
    cursor.execute(sql, (product["name"], product["price"]))

conn.commit() # 테이블의 데이터 수정되는 INSERT/UPDATE/DELETE는 실행 후 반드시 commit되어야만 최종 반영됨

print("10개의 상품이 추가되었습니다.")
```

```python
sql = "SELECT * FROM product"

cursor.execute(sql)

rows = cursor.fetchall() # fetchall()은 모든 데이터를 가져온다.

for row in rows:
    print(row) # 한 행은 튜플(tuple)로 반환된다.
```

```python
#id가 3인 모니터의 price null -> 150000으로 수정
update_price_val=150000
update_id=3

sql="""
update product
set price= %s
where id=%s
"""

cursor.execute(sql,(update_price_val,update_id))

conn.commit()

print("수정완료")
```

```python
try:
    sql=""" 
    delete from product
    where price is null
    """
#db null값을 연산할때는 db에서 제공하는 null연산자만을 사용해야한다.
# =, != 사용안됨

    cursor.execute(sql)
#select 실행 후 cursor에 반환된 결과셋을 가진다
#insert/update/delete는 cursor에 반환된 실행 행의 수를 가진다.
    print(f"{cursor.rowcount}행 삭제 성공")
    conn.commit()
except:
    print("db 예외 발생")
finally:
    cursor.close()
    conn.close()

    print("연결종료")
```
